Questão 21

```python
def obter_cidades():
    cidades = []

    while len(cidades) < 6:
        cidade = input(f"Digite o nome da cidade {len(cidades) + 1}: ").strip()

        if cidade.lower() in [c.lower() for c in cidades]:
            print("Cidade repetida! Digite um nome diferente.\n")
            continue

        cidades.append(cidade)

    return cidades


def concatenar_dados(cidades):
    print("\nDeseja concatenar informações a alguma cidade?")
    print("Digite o número referente ou 0 para pular.")

    for i, cidade in enumerate(cidades, start=1):
        print(f"{i} - {cidade}")

    opcao = input("Escolha: ")

    if opcao.isdigit():
        opcao = int(opcao)
    else:
        return cidades

    if opcao == 0:
        return cidades

    if 1 <= opcao <= len(cidades):
        extra = input("Digite o que deseja concatenar ao nome da cidade: ")
        cidades[opcao - 1] += " " + extra
        print("Concatenado com sucesso!\n")

    return cidades


def cidade_mais_extensa(cidades):
    mais_longa = max(cidades, key=len)
    return mais_longa

cidades = obter_cidades()
cidades = concatenar_dados(cidades)
maior_nome = cidade_mais_extensa(cidades)

print("\n========== RESULTADO ==========")
print("Cidades inseridas:")
for c in cidades:
    print("-", c)

print(f"\nCidade com nome mais extenso: {maior_nome} ( {len(maior_nome)} caracteres )")
```

Questão 33

```python
def calcular_imc(peso, altura):
    return peso / (altura ** 2)

pessoas = {}

while True:
    print("\n--- Cadastro de Pessoa ---")
    nome = input("Nome: ")
    altura = float(input("Altura (em metros): "))
    peso = float(input("Peso (em kg): "))
    cpf = input("CPF: ")
    sexo = input("Sexo (M/F): ")

    pessoas[cpf] = {
        "nome": nome,
        "altura": altura,
        "peso": peso,
        "sexo": sexo
    }

    opcao = input("\nDeseja cadastrar outra pessoa? (S/N): ").strip().upper()
    if opcao == "N":
        break

print("\n--- Consulta por CPF ---")
cpf_busca = input("Digite o CPF que deseja buscar: ")

if cpf_busca in pessoas:
    pessoa = pessoas[cpf_busca]
    imc = calcular_imc(pessoa["peso"], pessoa["altura"])

    print("\nPessoa encontrada:")
    print(f"Nome: {pessoa['nome']}")
    print(f"Altura: {pessoa['altura']} m")
    print(f"Peso: {pessoa['peso']} kg")
    print(f"Sexo: {pessoa['sexo']}")
    print(f"IMC: {imc:.2f}")

else:
    print("\nNenhuma pessoa encontrada com esse CPF.")
```
